## **TRAINING AND SPLITTING DATA**

```python
features=['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
x=df.loc[:,features]
y=df.loc[:,['Species']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

## **SUPPORT VECTOR MACHINE LEARANING ALGORITHM**

```python
from sklearn.svm import SVC
```

```python
task1 =  SVC()
task1.fit(x_train , y_train)
```

```python
predic=task1.predict(x_test)
predic
```

## **ACCURACY PREDICTION**

```python
from sklearn import metrics
Iris_svm = metrics.accuracy_score(predic,y_test)
print('The accuracy of the SVM is:', Iris_svm)
```
